Segmentasi Citra untuk Mendeteksi Mata Katarak dengan Metode Multithresholding


```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from skimage.color import rgb2hsv

rgb_img = io.imread('https://images.theconversation.com/files/281991/original/file-20190701-105176-h9f4wx.jpg?ixlib=rb-1.1.0&rect=0%2C186%2C2596%2C1296&q=45&auto=format&w=1356&h=668&fit=crop')

hsv_hsv = rgb2hsv(rgb_img)
hue_img = hsv_hsv[:,:,2]

fig, ([ax0, ax1]) = plt.subplots(1,2,figsize=(20,20))

ax0.imshow(hue_img)
ax1.imshow(hsv_hsv)

plt.show()
```

```python
import cv2
import numpy as np
from skimage import io

image = io.imread('https://images.theconversation.com/files/281991/original/file-20190701-105176-h9f4wx.jpg?ixlib=rb-1.1.0&rect=0%2C186%2C2596%2C1296&q=45&auto=format&w=1356&h=668&fit=crop')

gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
io.imshow(image)

ret, thresh=cv2.threshold(gray,55,255,60)
contours, hierarchy=cv2.findContours(thresh.copy(),cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)
n=len(contours)-1
contours=sorted(contours,key=cv2.contourArea,reverse=False)[:n]
for c in contours:
  hull=cv2.convexHull(c)
cv2.drawContours(image,[hull],0,(0,255,0),2)
io.imshow(image)
io.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage import io

img = io.imread ('https://images.theconversation.com/files/281991/original/file-20190701-105176-h9f4wx.jpg?ixlib=rb-1.1.0&rect=0%2C186%2C2596%2C1296&q=45&auto=format&w=1356&h=668&fit=crop')

grid_HSV = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

lower = np.array([0,100,100])
upper = np.array([400,255,255])

mask = cv2.inRange(grid_HSV, lower, upper)
res = cv2.bitwise_and(img, grid_HSV, mask=mask)

fig, ax = plt.subplots(1,4, figsize = (20,8))
ax[0].imshow(img)
ax[1].imshow(grid_HSV)
ax[2].imshow(mask)
ax[3].imshow(res)
plt.show()

```
